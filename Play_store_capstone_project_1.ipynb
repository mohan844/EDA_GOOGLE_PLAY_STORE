###5.Size:

```python
# Value counts:
df['Size'].value_counts()
```

**Size represents the size of App in Mb or Kb, there is one more label 'Varies with device'.Now we have to convert data into float. For 'Varies with device' label replace it with average size with the respective category.**

```python
# Function to convert object to float:
def remove_m(micro_byte):
  if micro_byte[-1] == 'M':
    return round(float(micro_byte.replace('M','')),1)
  elif micro_byte[-1] == 'k':
    return round(float(micro_byte.replace('k',''))/1024,1)
  else:
    return micro_byte
```

```python
df['Size'] = df['Size'].apply(remove_m)
```

**Replacing the Varies with Device label to average value of respective category**

```python
# Segregating the data with label Varies with Device:
size_df_1 = df.loc[df['Size']=='Varies with device']

# Segregating the data other than with theb label Varies with Device:
size_df_2= df.loc[df['Size']!='Varies with device']

# Group by the data by category and aggregate by mean size:
groupby_cat = size_df_2.groupby('Category')['Size'].mean().round(1)
```

```python
# Maping the data:
size_df_1['Size'] = size_df_1['Category'].map(groupby_cat)
```

```python
# Finally update the original DataFrame:
df.update(size_df_1)
```

```python
# Round the data upto one decimal Point:
df['Size'] = df['Size'].apply(lambda z: round(float(z),1))
```

```python
df.head()
```

```python
# Creating a list of list which contain all genres by using .tolist():
list_of_lists=df['Genres'].tolist()

# Removing the innerlist , so there is only single list:
single_list = [item for sublist in list_of_lists for item in sublist]

# Removing the duplicate Genres in single_list so in last we have only unique genres by using set:
unique_genres = list(set(single_list))
```

```python
unique_genres
```

```python
print(f'The total unique Genres are : {len(unique_genres)}')
```

```python
# Creating a dictionary which contains Genres as Key and value as total count:
genres_feq = {}
for genre in unique_genres:
  count = single_list.count(genre)
  genres_feq[genre] = count
```

```python
# Creating a another Dictionary which we can pass it into DataFrame:
main_dict = {'Genres':list(genres_feq.keys()), 'Total no of Frequency':list(genres_feq.values())}

# Creating a DataFrame:
genres_df = pd.DataFrame(main_dict)
```

##3. Category and Rating:

```python
# Creating a fuction who divide the ratting into three bins:
def rating_binning(rating_value):
  if rating_value > 3.7:
    return 'High Rated'
  elif rating_value > 2.3:
    return 'Average Rated'
  else:
    return 'Low Rated'
```

```python
df['Rating_Type'] = df['Rating'].apply(rating_binning)
```

```python
# Creating a new dataframe by using groupby by type of rated and calculate percent of high rated rating in respective category:
cat_rating_group = df.groupby(['Category','Rating_Type'])['index'].count().unstack().fillna(0)
cat_rating_group['High_rated_percent'] = (cat_rating_group['High Rated'] /(cat_rating_group['High Rated'] + cat_rating_group['Average Rated'] + cat_rating_group['Low Rated']) * 100).round(2)
cat_rating_group.sort_values(by='High_rated_percent',ascending = False)
```

```python
# Ploting Subplots Boxplot and bar plot high rating apps of different category:
fig,axes = plt.subplots(1,2,figsize=(35,10))
# Box Plot
sns.boxplot(data=df, x='Category', y='Rating', ax=axes[0])
# Average price Plot:
sns.barplot(data=cat_rating_group, x=cat_rating_group.index, y='High_rated_percent',ax=axes[1])
# Labeling the Graphs:
axes[0].tick_params(axis='x', labelrotation=90)
axes[0].set_title('Box plot')
axes[1].tick_params(axis='x', labelrotation=90)
axes[1].set_title('High Rated rating % ')
```

```python
# Ploting the data:
cat_summary
```

```python
# Correlation Heat Maps Heat maps :
corr_df = cat_summary.corr()
corr_df
plt.figure(figsize=(12,10))
sns.heatmap(corr_df, cmap="crest",annot=True)
```

**Conclusion:**
The correlation coefficient measures the strength and direction of the linear relationship between two variables. A coefficient of 1 means a perfect positive correlation, -1 means a perfect negative correlation, and 0 means no correlation at all.

*   total_rating and total_reviews have a moderate positive correlation of 0.54, meaning that as one variable increases, the other tends to increase as well, but not necessarily at the same rate.
*   total_rating and Installs have a moderate positive correlation of 0.52, which suggests that apps with higher ratings tend to have more installs.
